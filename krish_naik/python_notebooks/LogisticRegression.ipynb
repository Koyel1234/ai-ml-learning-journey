# Logistic Regression: One vs Rest (Chris Albon)
On their own, Logistic Regressions are only binary classifiers, meaning they cannot handle target vectors with more than two classes. However, there are clever extensions to Logistic Regression to do just that.In one-vs-rest Logistic Regression (OVR) a seperate model is trained for each class predicted whether whether an observation is that class or not (thus making it a binary problem). It assumes that each classification problem (e.g. class 0 or not) is independent.

```python
#load libraries
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
#Load iris flower data
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# standardize features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
#Create One-vs-Rest Logistic Regression
clf = LogisticRegression(random_state=0, multi_class='ovr')
```

```python
#Train One-vs-Rest Logistic Regression
#train model
model = clf.fit(X_std, y)
```

```python
#create previously unseen (new) observation
new_observation = [[0.5, 0.5, 0.5, 0.5]]
```

```python
#predict observation's class
model.predict(new_observation)
```

```python
#view probability observation in each class
#view predicted probability
model.predict_proba(new_observation)
```
