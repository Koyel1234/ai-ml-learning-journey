## Classes
A structure containing varoiables and functions.
Ex. Create a Shopping List class that can add and subtract items and their quantities

```python
class ShoppingList:
    def __init__ (self):
        self.listdict={}  #empty list is created
        
    def addItem(self,key,val):
        if key in self.listdict:
            self.listdict[key]=self.listdict[key]+val
        else:
            self.listdict[key]=val
    
    def subtractItem(self,key,val):
        if key in self.listdict:
            self.listdict[key]=self.listdict[key]-val
            
    def removeItem(self,key):
        if key in self.listdict:
            del(self.listdict[key])
            
    def printList(self):
        print(self.listdict)
        
        
        
# there can be one ShoppoingList for fruits, another ShppingList class for clothes, another for accessories.....        
```

```python
sl1=ShoppingList()
sl1.addItem("Banana",12)
sl1.addItem("Apple",34)
sl1.printList()
sl1.subtractItem("Banana",2)
sl1.printList()
sl1.removeItem("Apple")
sl1.printList()
sl1.subtractItem("Guava",2)
sl1.printList()
```

```python
sl2=ShoppingList()
sl2.printList()
sl2.addItem("notebook",4)
sl2.addItem("copy",5)
sl2.printList()
sl2.addItem("pen",8)
sl2.printList()
sl2.removeItem("pen")
sl2.subtractItem("copy",1)
sl2.printList()
```

sl1 and sl2 are two instances of the class ShoppingList

### Matrix Concatenation

```python
print(A.T.shape, B.shape)
C = np.concatenate((A.T,B), axis=1)  #axis=1 means column wise
D = np.concatenate((A.T,B), axis=0)   #axis=0 means row wise
print(C)
print(D)
```
