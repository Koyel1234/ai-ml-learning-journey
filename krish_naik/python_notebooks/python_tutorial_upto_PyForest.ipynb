### Boolean Variables

Boolean values are the two constant objects False and True.

They are used to represent truth values (other values can also be considered false or true).

In numeric contexts (for example, when used as the argument to an arithmatic operator), they behave like the integers 0 and 1, respectively.

The built-in function bool() can be used to cast any value to a Boolean, if the value can be interpreted as a truth value.

They are written as False and True, respectively.

# List Comprehension
List comprehension provide a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The expressions can be anything, meaning you can put in all kinds of objects in lists.

```python
lst1=[]
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1    
```

```python
lst_square([1,2,3,4,5,6])
```

```python
lst=[1,2,3,4,5,6]
[i*i for i in lst]
```
