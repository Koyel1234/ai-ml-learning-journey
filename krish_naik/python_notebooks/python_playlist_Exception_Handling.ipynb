# Python Assert 
Python provides the assert statement to check if a given logical expression is true or false. Programme execution proceeds only if the expression is true and raises the AssertionError when it is false. The following code shows the usage of the assert statement. 

```python
10>10
```

```python
10>=10
```

```python
num=10
assert num>=10  #got executed successfully
```

```python
num=10
assert num>10  #raises assertion error
```

```python
try:
    num=int(input("Enter the number"))
    assert num%2==0
    print("The number is even")
except AssertionError:
    print("Please enter even number")
    
## give even input    
```

```python
try:
    num=int(input("Enter the number"))
    assert num%2==0
    print("The number is even")
except AssertionError:
    print("Please enter even number")
    
## give odd input    
```
