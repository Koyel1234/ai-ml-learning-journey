```python
from PyPDF2 import PdfFileReader,PdfFileWriter
```

```python
#open the current pdf
file_pdf=PdfFileReader('Thapa_Bimal.pdf')
#object for pdf writer
out_pdf=PdfFileWriter()
```

```python
file_pdf
```

```python
file_pdf.numPages
```

#  Include all pages

```python
for i in range(file_pdf.numPages):
    page_details=file_pdf.getPage(i)
    #add to the output page
    out_pdf.addPage(page_details)
```

```python
password="Koyel@1234"

out_pdf.encrypt(password)
```

```python
with open("encryptedpdf_Thapa_Bimal.pdf","wb") as filename:
    out_pdf.write(filename)
```

# Include all pages except last 5 pages

```python
for i in range(file_pdf.numPages-5):
    page_details=file_pdf.getPage(i)
    #add to the output page
    out_pdf.addPage(page_details)
```

```python
password_n="Koyel@12345"

out_pdf.encrypt(password_n)
```

```python
with open("encryptedpdf_new_Thapa_Bimal.pdf","wb") as filename:
    out_pdf.write(filename)
```
