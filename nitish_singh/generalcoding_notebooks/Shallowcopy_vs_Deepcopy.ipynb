```python
## =, copy(), deepcopy()
lst1 = [1,2,3,4]
lst2 = lst1
```

```python
lst1, lst2
```

```python
print(lst2[1])
lst2[1] =1000
lst1, lst2 # both variables lst1 and lst2 sharing same memory loc
```

```python
## copy operation
lst1 = [1,2,3,4]
lst2 = lst1.copy() # shallow copy # different memory loc
lst1, lst2
```

```python
lst2[1] = 1000
lst1, lst2
```

```python
lst3 = [[1,2,3,4], [5,6,7,8]]
lst4 = lst3.copy()
```

```python
lst3, lst4
```

```python
lst3[1][1] = 1000
lst3, lst4
```

```python
lst3.append([2,3,4,6])
lst3, lst4
```

```python
## deepcopy
import copy
lst5 = [1,2,3,4]
lst6 = copy.deepcopy(lst5)
lst5, lst6
```

```python
lst6[1] = 1000
lst5, lst6 # for normal list deepcopy and shallow copy are same
```

```python
lst7 = [[1,2,3,4], [5,6,7,8]]
lst8 = copy.deepcopy(lst7) # copy.copy() is shallow copy as .copy() earlier
lst7, lst8
```

```python
lst8[1][1] = 1000
lst7, lst8
```
