```python
lst = [1,2,3,4,5,6,7]

def evenoddsum(lst):
    even_sum = 0
    odd_sum = 0
    for i in lst:
        if i%2 == 0:
            even_sum += i
        else:
            odd_sum += i
    return even_sum, odd_sum
```

```python
evenoddsum(lst)
```

# List Comprehension

List comprehensions provides a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The expressions can be anything, meaning you can put in all kinds of objects in lists.

```python
lst1 = []
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
```

```python
lst_square([1,2,3,4,5,6,7])
```

```python
# list comprehension
lst = [1,2,3,4,5,6,7]
lst2=[i*i for i in lst]
lst2
```

```python
lst3=[i*i for i in lst if i%2==0] # square of even numbers
lst3
```

```python
lst4=[i*i for i in lst if i%2!=0] # square of odd numbers
lst4
```

# Python Iterables vs Iterators

```python
# list is iterable, all elements of list while looping is geeting stored into memory
lst = [1,2,3,4,5,6,7]

for i in lst: 
    print(i)
```

```python
# whenever we will convert a list into an iterator, all elements will not be allocated to memory, 
# only when we'll call inbuild method next() then only that particular element will be allocated to memory
# internally one pointer is working which will point particular value at a time and that will be allocated to memory
lst_it = iter(lst)
```

```python
next(iter(lst_it))
```

```python
next(iter(lst_it))
```

```python
next(iter(lst_it))
```

```python
next(iter(lst_it))
```

```python
print(next(iter(lst_it)))
```

```python
print(next(iter(lst_it)))
print(next(iter(lst_it)))
print(next(iter(lst_it)))
print(next(iter(lst_it)))
```

```python
lst_it1 = iter(lst)
for i in lst_it1:
    print(i)
```

```python
lst_it2 = iter(lst)
print(next(lst_it2))
print("starting for loop")
for i in lst_it2: # by default for loop will come out after last loop, it will not through error as we will get next() function, StopIteration exception is already handled in for loop
    print(i)
```

1. Parse python Expression
2. Compile into a byte code
3. Evaluate the python expression
4. It will return the result
