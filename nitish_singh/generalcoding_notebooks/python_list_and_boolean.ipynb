## Boolean Variables
Boolean values are the two constant objects **True** and **False**.

They are used to represent truth values (other values can also be considered false or true).

In numeric contexts, for example when used as an argument in an arithmatic operator, they behave like integers 0 and 1, respectively.

The built-in function bool() can be used to cast any value to a Boolean, if the value can be interpreted as a truth value.

They are written as **False** and **True** respectively.

```python
a = bool()
a # default value False
```

```python
print(True, False)
```

```python
type(True), type(False)
```

```python
my_name = "Koyel Pramanick"

print(my_name.isalnum())
```

```python
my_name = "Koyel Pramanick13"

print(my_name.isalnum())
```

```python
my_name = "KoyelPramanick13"

print(my_name.isalnum()) # blank space should not come in alnum, alnum mean either alphabet or numbers
```

```python
my_name = "Koyel Pramanick"

print(my_name.isalnum()) # check if all char are numbers
print(my_name.isalpha()) # check if all char in the string are alphabets 
print(my_name.isdigit()) # test if string contains digit 
print(my_name.istitle()) # test if string contains title words
print(my_name.isupper()) # test if string contains upper case
print(my_name.islower()) # test if string contains lower case 
print(my_name.isspace()) # test if string is whitespace string 
print(my_name.endswith('k')) # test if string ends with 'k'
print(my_name.startswith('K')) # test if string starts with 'K'
```

## List 
A list is a data structure in Python that is mutable or changeable, ordered sequence of elements. Each element or value that is inside of a list is called an item. Just as strings are defined as characters between quotes, lists are defined by having values between square brackets [].

```python
type([])
```

```python
lst_example = [] # blank list
type(lst_example)
```

```python
lst_example1 = list()
type(lst_example1)
```

```python
lst_example2 = ['Mathematics', 'science', 1, 2, "ABCD"]
lst_example2
```

```python
# append, extend

lst_example2.append('Biology') # will add as another normal element for adding one element 
lst_example2
```

```python
lst_example2.append(4)
lst_example2
```

```python
lst_example2.append(['Biology']) # if will try to add more than one one element or specifically a list, it will be added as list not seperate elements
lst_example2
```

```python
lst_example2.extend(['A', 'B']) # will add all elements of list as seperate elements
lst_example2
```

```python
# Indexing
lst_example2[0]
```

```python
lst_example2[-1] # list indexing works from both end
```

```python
lst_example2[:]
```

```python
lst_example2[1:]
```

```python
lst_example2[:-1]
```

```python
lst_example2[2:5] # 2nd to 4th place , list index is not right inclusive
```

```python
# insert
lst_example2.insert(1, "Check")
lst_example2
```

```python
# operations on list
lst_example3 = [1,2,3,4,5,6]
print(sum(lst_example3))
print(min(lst_example3))
print(max(lst_example3))
```

```python
# pop() method - by default will remove last element
# if index is mentioned, then will remove element of that index
print(lst_example3.pop()) # without index one is similar to stack
print(lst_example3)
```

```python
print(lst_example3.pop(2))
print(lst_example3)
```

```python
lst_example4 = [1, 1, 2, 3, 4]
print(lst_example4.count(1))
print(lst_example4.count(10))
```

```python
len(lst_example4)
```

```python
# index : index(value for which index to find, start index, end index) : start and end indexes are not mandatory
# index function will return index of occurance (in case of single occurances) or first occurance (in case of multiple occurances) (within range if start and end indexes are provided)
print(lst_example4.index(1))
print(lst_example4.index(1, 0, 4))
print(lst_example4.index(1, 1, 4))
print(lst_example4.index(1, 2, 4)) # will get error
```

```python
print(lst_example4*2) # whole list will be extended twice
```

## Sets

A Set is an unordered collection data type that is iterable, mutable and has no duplicate elements. Python's set class represent the mathematical notion of a set. This is based on a data structure known as hash table.

```python
set_var = set() # empty set
print(set_var)
print(type(set_var))
```

```python
set_var = {"I", "J", "K"}
print(set_var)
print(type(set_var))
```

```python
set_var = {"I", "J", "K", "I"} # remove duplicates
print(set_var)
print(type(set_var))
```

```python
set_var = {"I", "K", "J", "I"} # following alphabetical order
print(set_var)
print(type(set_var))
```

```python
## set does not support indexing
set_var[0]
```

```python
set_var["I"]
```

```python
# in built functions
set_var.add("L")
set_var
```

```python
# difference : elements which are in set2 not in set1
# will not update set2 or set1
set1 = {"a", "b", "c"}
set2 = {"a", "b", "c", "d"}
print(set2.difference(set1))
print(set2, set1)
```

```python
# difference_update : elements which are in set2 not in set1
# will update set2 or set1
set1 = {"a", "b", "c"}
set2 = {"a", "b", "c", "d"}
print(set2.difference_update(set1))
print(set2, set1)
```

```python
# intersection : elements which are in both set2 and set1
# will not update set2 or set1
set1 = {"a", "b", "c"}
set2 = {"a", "b", "c", "d"}
print(set2.intersection(set1))
print(set2, set1)
```

```python
# intersection_update : elements which are in both set2 and set1
# will update set2 or set1
set1 = {"a", "b", "c"}
set2 = {"a", "b", "c", "d"}
print(set2.intersection_update(set1))
print(set2, set1)
```

## Dictionary

A dictionary is a collection which is unordered, changeable and indexed. In Python, dictionaries are written within curly brackets and they have keys and values.

```python
dict = {} # empty dictionary

print(dict)
print(type(dict))
```

```python
dict = {1,2,3,4}
type(dict) # this is set not dictionary
```

```python
dict = {"Car1" : "Audi", "Car2" : "BMW", "Car3" : "Mercidies Benz"}
print(dict)
print(type(dict))
```

```python
# access the item values by keys
dict["Car1"]
```

```python
# we can even loop through dictionary keys, during for loop by default keys will be printed
for i in dict:
    print(i)
```

```python
# we can loop through dictionary values
for i in dict.values():
    print(i)
```

```python
for i in dict.keys():
    print(i)
```

```python
for i in dict.items():
    print(i)
```

```python
# add new item in dictionary
dict["Car4"] = "Audi 2.0"
dict # new item added
```

```python
# update old item
dict["Car3"] = "Maruti"
dict
```

```python
## nested dictionaries
car1_desc = {"model" : "Audi", "year" : 2000}
car2_desc = {"model" : "BMW", "year" : 2002, "Colour" : "black"}
car3_desc = {"model" : "Maruti"}

cars_desc = {"car1" : car1_desc, "car2" : car2_desc, "car3" : car3_desc}
cars_desc
```

```python
type(cars_desc)
```

```python
print(cars_desc['car1'])
print(cars_desc['car1']['model'])
```

## Tuple

Tuples are immutable.

```python
tuple_ex = tuple() # empty tuple
print(tuple_ex)
print(type(tuple_ex))
```

```python
t_ex = ("A", "B", "C")
print(t_ex)
print(type(t_ex))
```

```python
t_ex[0]
```

```python
t_ex[0] = "D" # immutable
```

```python
t_ex1 = ("a", "b", "c", "c") # will not remove duplicates
t_ex1
```

```python
# inbuilt function
print(t_ex1.count("a"))
print(t_ex1.count("c"))
```

```python
t_ex2 = (1,2,3,4)
print(sum(t_ex2))
print(min(t_ex2)) 
print(max(t_ex2))
```

```python
print(t_ex1.index("a"))
print(t_ex1.index("c")) # will return 1st occurance as multiple occurances present
```
