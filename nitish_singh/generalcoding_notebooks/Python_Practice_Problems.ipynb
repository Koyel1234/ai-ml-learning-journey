```python
# User will input (2 numbers).Write a program to swap the numbers
def swap():
    num1 = int(input("Print first num : "))
    num2 = int(input("Print second num : "))
    print("first, second before swapping : ", num1, num2)
    num3 = num1
    num1 = num2
    num2 = num3
    print("first, second after swapping : ", num1, num2)

swap()
```

```python
# Write a program that will reverse a four digit number.Also it checks whether the reverse is true.
def reverse_num_check(num:int, reverse_num):
    # assert type(num)==int, 'Input is not integer'
    rev = int(str(num)[::-1])
    if rev == reverse_num:
        flag = True
    else:
        flag = False
    return (rev, flag)
reverse_num_check(1234,4321)
```

```python
# Write a program to find the euclidean distance between two coordinates. ((distance between (x1,y1), (x2,y2) - rootover((x1-x2)^2 + (y1-y2)^2)))
import math
def distance(*args):
    l = len(args[0])
    d = 0
    for i in range(l):
        d += (args[1][i] - args[0][i])**2 
    return math.sqrt(d)

distance((1,1,1), (0,0,0))
```

```python
# Write a program that take a user input of three angles and will find out whether it can form a triangle or not.
def form_triangle(a1,a2,a3):
    if a1+a2+a3 == 180:
        return True
    else:
        return False

form_triangle(40,45,95)
```

```python
# Write a program to print the following pattern
# *
# **
# ***
# ****
# *****
def pat1(n):
    for i in range(1,n+1):
        print('*'*i)

pat1(5)
```

```python
# Write a program to print the following pattern
# *
# **
# ***
# **
# *
def pat2(n): # either add check cond for n should be odd always or for even n, in last line there will be 2 stars instead of 1
    t = n//2
    for i in range(1, t+1):
        print('*'*i)
    print('*'*(t+1))
    # for i in reversed(range(1, t+1)):
    #     print('*'*i)
    for i in range(t, 0, -1):
        print('*'*i)

pat2(5)
```

```python
# Write  a program to print the following pattern
#         *
#       * * *
#     * * * * *
#    * * * * * * *
# * * * * * * * * *
def pat3(n):
    for i,j in zip(range(1, n+1), range(n-1,-1,-1)):
        l = 2*i - 1
        print('  '*j+'* '*l)

pat3(9)
```
