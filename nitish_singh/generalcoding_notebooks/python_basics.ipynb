```python
# counting unique vowels in a given word
vowels = ['a', 'e', 'i', 'o', 'u']

def count_unique_vowel(word):
    c = 0
    word_lower = word.lower()
    for char in vowels:
        if char in word_lower:
            c += 1
    print(c)
```

```python
count_unique_vowel('AEIOUa')
```

```python
# counting vowels in a given word with repeatation
vowels = ['a', 'e', 'i', 'o', 'u']

def count_repeate_vowel(word):
    c = 0
    word_lower = word.lower()
    for char in word_lower:
        if char in vowels:
            c += 1
    print(c)
```

```python
count_repeate_vowel('AEIOUa')
```

```python
# counting repeate consonants in a given word
vowels = ['a', 'e', 'i', 'o', 'u']

def count_repeate_consonants(word):
    c = 0
    word_lower = word.lower()
    for char in word_lower:
        if char not in vowels:
            c += 1
            print(char)
    print(c)

count_repeate_consonants('AmIooRrf')
```

```python
# counting unique consonants in a given word
vowels = ['a', 'e', 'i', 'o', 'u']

def count_unique_consonants(word):
    l = []
    word_lower = word.lower()
    for char in word_lower:
        if char not in vowels:
            print(char)
            l.append(char)
    print(len(set(l)))

count_unique_consonants('AmIooRrf')
```

```python
# finding maximum number in a list without using max function
def max_num(lst):
    max = lst[0]
    for l in range(1, len(lst)):
        if lst[l] > max :
            max = lst[l]
    print(max)

max_num([2,67,89,2,3,9])
```

```python
# finding minimum number in a list without using min function
def min_num(lst):
    min = lst[0]
    for l in range(1, len(lst)):
        if lst[l] < min :
            min = lst[l]
    print(min)

min_num([2,67,89,2,3,9])
```

```python
# comparing two strings for anagrams - space is allowed in new word, number of repeatation should match
def anagrams(w1,w2):
    w1 = w1.replace(" ","").lower()
    w2 = w2.replace(" ","").lower()
    if sorted(w1) == sorted(w2):
        print(f"Yes, '{w1}' and '{w2}' are anagrams.")
    else:
        print(f"No, '{w1}' and '{w2}' are not anagrams.")
```

```python
# checking for palindrome using extended slicing technique
def check_palindrome(str):
    str_lower = str.lower()
    if str_lower == str_lower[::-1]:
        print(f"Yes, '{str}' is palindrome.")
    else:
        print(f"No, '{str}' is not palindrome.")
```

# Decision Control Statements 

```python
# correct email - campusx@gmail.com
# password = 1234

email = input("Enter your email : ")
if '@' in email:
    password = input("Enter your password : ")
    if email == "campusx@gmail.com" and password == '1234':
        print("Welcome")
    elif email == 'campusx@gmail.com' and password != '1234':
        print("Password Incorrect")
        password = input("Enter correct password: ")
        if password == '1234':
            print("Finally correct")
        else:
            print("Still Incorrect")
    else:
        print("Incorrect credentials")
else:
    print("Enter correct email")
```
