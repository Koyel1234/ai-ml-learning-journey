## Stemming
Stemming is the process of reducing a word to its word stem that affixes and prefixes to the roots of words known as lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP)

```python
## classification Problem
## Comments of product is a positive review or negative review
## reviews --> eating/eat/eaten : variants of eat: root or stem word is 'eat'
## going, gone, goes : root word - 'go'
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalize"]
```

```python
## Porter Stemmer
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

```python
## for some words stemming can change meaning
```

## RegexpStemmer class

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expressions Stemmer Algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example:

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
reg_stemmer1 = RegexpStemmer('ing|s$|e$|able$', min=4)
reg_stemmer1.stem('ingeating')
```

## Wordnet Lemmatizer
Lemmatization technique is like stemming. The output we will get after lemmatization is called 'lemma', which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example.

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
'''
POS - 
noun - n
verb - v
adjective - a
adverb - r
'''
lemmatizer.lemmatize("going", pos='r') # default pos = 'n'
```

```python
lemmatizer.lemmatize("going", pos='v')
```

```python
lemmatizer.lemmatize("going", pos='a')
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word))
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word, pos="v"))
```

```python
lemmatizer.lemmatize("goes", pos="v")
```

```python
lemmatizer.lemmatize("fairly", pos="r")
```
