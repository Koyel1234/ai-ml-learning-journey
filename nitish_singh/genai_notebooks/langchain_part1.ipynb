```python
from langchain_openai import OpenAIEmbeddings
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
embedding = OpenAIEmbeddings(
    model = 'text-embedding-3-large',
    dimensions = 32
)

documents = ["Delhi is the capital of India.",
             "Kolkata is the capital of West Bengal.",
             "Paris is the capital of France."
            ]
```

```python
query = "Tell something about Delhi."
```

```python
doc_embeddings = embedding.embed_documents(documents)
query_embedding = embedding.embed_query(query)
```

```python
scores = cosine_similarity([query_embedding], doc_embeddings)[0] # to get list
```

```python
index, score = sorted(list(enumerate(scores)), key=lambda x: x[1])[-1]
```

```python
print(query)
print(documents[index])
print('similarity score: ', score)
```

```python
# use structured tool
from langchain.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(description='The first number to multiply')
    b: int = Field(description='The second number to multiply')

def multiply_func(a: int, b: int) -> int:
    return a * b

multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    name = 'multiply',
    description = 'Multiply two numbers',
    args_schema = MultiplyInput
)
```

```python
result = multiply_tool.invoke({"a":5, "b":6})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
# using BaseTool class
from langchain.tools import BaseTool
from typing import Type

# arg schema using pydantic
class MultiplyInput(BaseModel):
    a: int = Field(description='The first number to multiply')
    b: int = Field(description='The second number to multiply')

class MultiplyTool(BaseTool):
    name: str = 'multiply'
    description: str = 'Multiply two numbers'
    args_schema: Type[BaseTool] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
result = multiply_tool.invoke({"a":7, "b":8})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
# Toolkits
from langchain_core.tools import tool

# custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b


class MathToolkit:
    def get_tools(self):
        return [add, multiply]
    

toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```

```python
## injected tool argument
from langchain_core.tools import InjectedToolArg
from typing import Annotated


@tool
def get_conversion_factor(base_currency: str, target_currency: str) -> float:
    """This function fetches the currency conversion factor from base_currency to target_currency using an external API."""
    response = requests.get(f"https://api.exchangerate-api.com/v4/latest/{base_currency}") # change this url according shown in video or see if this is working
    data = response.json()
    return data['rates'].get(target_currency, None)

@tool
def convert(base_currency_value: float, conversion_rate: Annotated[float, InjectedToolArg]) -> float:
    """Given a currency conversion rate this function calculates the target currency value from a given base currency value."""
    return base_currency_value * conversion_rate



llm = ChatOpenAI()
llm_with_tools = llm.bind_tools([get_conversion_factor, convert])
messages = [HumanMessage('What is the conversion rate from USD to INR and based on that can you convert 10 usd to inr?')]
ai_message = llm_with_tools.invoke(messages)
ai_message.tools_calls # conversion rate not set
```

```python
import json
for tool_call in ai_message.tools_calls:
    print(tool_call)
    # execute the 1st tool and get the value of conversion rate
    if tool_call.tool_name == 'get_conversion_factor':
        tool_message1 = get_conversion_factor.invoke(tool_call)
        # fetch conversion rate
        converion_rate = json.loads(tool_message1.content)['converion_rate']
        # append this tool message to meassage list
        messages.append(tool_message1)
    # execute the 2nd tool using conversion rate from tool 1
    if tool_call.tool_name == 'convert':
        # fetch the current arg
        tool_call['args']['converion_rate'] = converion_rate
        tool_message2 = convert.invoke(tool_call)
        messages.append(tool_message2)
```
