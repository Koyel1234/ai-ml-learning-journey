```python
data1 = ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11')

data1
```

```python
pd.read_csv(StringIO(data1))
```

```python
pd.read_csv(StringIO(data1)).dtypes 
# as there is nan in col d, so it converted to float,  column type can't be int if there is null, it will be float
```

```python
print(pd.read_csv(StringIO(data1), dtype=float))
print(pd.read_csv(StringIO(data1), dtype=float).dtypes)
print(pd.read_csv(StringIO(data1), dtype=float)['a'].dtype)
```

```python
print(pd.read_csv(StringIO(data1), dtype=object))
print(pd.read_csv(StringIO(data1), dtype=object).dtypes)
print(type(pd.read_csv(StringIO(data1), dtype=str)['a'][1]))
```

```python
print(pd.read_csv(StringIO(data1), dtype=str))
print(pd.read_csv(StringIO(data1), dtype=str).dtypes)
print(type(pd.read_csv(StringIO(data1), dtype=str)['a'][1]))
```

```python
print(pd.read_csv(StringIO(data1), dtype={'a':'Int64', 'b': float, 'c':float}))
print(pd.read_csv(StringIO(data1), dtype={'a':'Int64', 'b': float, 'c':float}).dtypes)
print('\n')

print(pd.read_csv(StringIO(data1), dtype={'a':'Int64', 'b': int, 'c':float}))
print(pd.read_csv(StringIO(data1), dtype={'a':'Int64', 'b': int, 'c':float}).dtypes)
print(type(pd.read_csv(StringIO(data1), dtype={'a':'Int64', 'b': int, 'c':float})['a'][1]))
print(type(pd.read_csv(StringIO(data1), dtype={'a':'Int64', 'b': int, 'c':float})['b'][1]))
```

```python
# set index
data_idx = ('index,a,b,c\n' '1,apple,bat,5.7\n' '2,orange,cow,10')
display(pd.read_csv(StringIO(data_idx)))
display(pd.read_csv(StringIO(data_idx), index_col = 0)) # make 0th col as index, col indexing starts from 0
display(pd.read_csv(StringIO(data_idx), index_col = 1)) # make 1st col as index, col indexing starts from 1
```

```python
data_idx1 = ('a,b,c,d\n' '1,apple,bat,5.7\n' '2,orange,cow,10')
data_idx1
display(pd.read_csv(StringIO(data_idx1))) # not making 1st column as index
display(pd.read_csv(StringIO(data_idx1), index_col = False)) # not making 1st column as index
```

```python
# put comma after last item except 1st row
data_idx2 = ('a,b,c,d\n' '1,apple,bat,5.7,\n' '2,orange,cow,10,')
data_idx2
display(pd.read_csv(StringIO(data_idx2))) # making 1st column as index 
display(pd.read_csv(StringIO(data_idx2), index_col = False)) # not making 1st column as index
```

```python
# put comma after last item in 1st row as well
data_idx2 = ('a,b,c,d,\n' '1,apple,bat,5.7,\n' '2,orange,cow,10,')
data_idx2
display(pd.read_csv(StringIO(data_idx2))) # making 1st column as index 
display(pd.read_csv(StringIO(data_idx2), index_col = False)) # not making 1st column as index
```

```python
# quoting and escape characters, very useful in NLP

data = 'a,b"hello, \\"Bob\\", nice to see you",5'
pd.read_csv(StringIO(data), escapechar='\\')
```

```python
## Read JSON to CSV
# putting dictionary of title1 and title2 in list in title 
data = '{"employee_name" : "James", "email" : "james@gmail.com", "title" : [{"title1" : "Team Lead", "title2" : "Sr. Developper"}]}'
display(pd.read_json(data))
                    
# putting dictionary of title1 and title2 without list in title 
data = '{"employee_name" : "James", "email" : "james@gmail.com", "title" : {"title1" : "Team Lead", "title2" : "Sr. Developper"}}'
display(pd.read_json(data))

# putting dictionary of title1 and title2 without list in title 
data = '{"employee_name" : "James", "email" : {"email_primary" : "james@gmail.com", "email_secondary" : "james@gmail.com"}, "title" : {"title1" : "Team Lead", "title2" : "Sr. Developper"}}'
display(pd.read_json(data))
```

```python
data = '{"employee_name" : "James", "email" : "james@gmail.com", "title" : [{"title1" : "Team Lead", "title2" : "Sr. Developper"}]}'
data_json = pd.read_json(data)
display(data_json)

display(data_json.to_json())
display(data_json.to_json(orient='index'))
display(data_json.to_json(orient='records'))
```

```python
## Read table from web page
# match parameter can be regex as well as specific string
url_mobile_country_code = "https://en.wikipedia.org/wiki/Mobile_country_code" 
df1 = pd.read_html(io = url_mobile_country_code, match = "Country") # will return tables containing string "Country", 
df1
```

```python
df2= pd.read_html(io = url_mobile_country_code, match = "Brand") # will return tables containing string "Brand"
df2
```

```python
print(type(df1), len(df1))
print(type(df2), len(df2))
```

```python
df3= pd.read_html(io = url_mobile_country_code) # returning all tables contained in that specifi webpage
print(type(df3), len(df3))
display(df3)
```
