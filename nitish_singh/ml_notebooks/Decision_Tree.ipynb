```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
## independent and dependent features
X = df.iloc[:, :-1]
y= iris.target
```

```python
X
```

```python
y
```

```python
## tarin test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33 , random_state=42)
```

```python
X_train
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
## decisiontree 
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
# this is cart tree as all aplits are binary only
```

## Postprunning

```python
treemodel1 = DecisionTreeClassifier(max_depth=2)
treemodel1.fit(X_train, y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel1, filled=True)
```

```python
#prediction
y_pred = treemodel1.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score=accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```

## Preprunning

```python
parameter = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto', 'sqrt', 'log2']
    # 'ccp_alpha' : [1,2,3,4,5,6,7]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel2 = DecisionTreeClassifier()
cv = GridSearchCV(treemodel2, param_grid=parameter, cv=5, scoring='accuracy')
# in CV how to decide train and test? randomly? bcz if they will be in some order then may be in first segment only one type of calss will be there
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
y_pred = cv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
score
```

```python
print(classification_report(y_pred, y_test))
```
