## Matplotlib

Matplotlib is a plotting library for Python Programming Language and its numerical mathematical extension NumPy. It provides an object oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, wxPython, QT or GTK+ .

Some of the major Pros of Matplotlib:
- Generally easy to get started for simple plots.
- Support for custom labels and texts.
- Great control of every element in a figure.
- High quality output in many formats.
- Very customizable in general.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## plot examples
x = np.arange(0,10)
y = np.arange(11,21)

plt.scatter(x,y)
```

```python
plt.scatter(x,y, c= "r")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Main_Title")
plt.savefig("Test.png")
```

```python
## next example
y = x*x
plt.plot(x,y)
plt.show()
plt.plot(x,y,"r")
plt.show()
plt.plot(x,y,"ro")
plt.show()
plt.plot(x,y,"r+")
plt.show()
plt.plot(x,y,"r", linestyle="dashed")
plt.show()
plt.plot(x,y,"ro", linestyle="dashed", linewidth = 2, markersize = 12)
plt.show()
plt.plot(x,y,"ro", linestyle="dashed", linewidth = 0.5, markersize = 5)
# plt.show() # if we are keeping this plt.show() here, then plot will be executed and then axis labels will not be applied
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Main Title")
```

```python
## creating subplots
plt.subplot(2,2,1).set_title("t1")
plt.plot(x,y,"r--")
plt.xlabel("x1")
plt.subplot(2,2,2)
plt.plot(x,y,"go")
plt.subplot(2,2,4)
plt.plot(x,y,"b", linestyle = "dashed")
plt.xlabel("x1")
plt.xlabel("x1")
```

```python
## creating subplots
ax = plt.figure()
ax.suptitle("main title")
ax1 = plt.subplot(2,2,1)
ax1.plot(x,y,"r--")
plt.xlabel("x1")
ax1.set_title("t1")
plt.subplot(2,2,2)
plt.plot(x,y,"go")
plt.subplot(2,2,4)
plt.plot(x,y,"b", linestyle = "dashed")
plt.xlabel("x1")
plt.xlabel("x1")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
ax1.title.set_text('First Plot')
ax2.title.set_text('Second Plot')
ax3.title.set_text('Third Plot')
ax4.title.set_text('Fourth Plot')
plt.show()
```

```python
# compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sin Wave Form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
# compute x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# setup subplot grids with height = 2 and width = 1

# first subplot
plt.subplot(2,1,1)
plt.plot(x, y_sin)
plt.title("sin curve")

# second subplot
plt.subplot(2,1,2)
plt.plot(x, y_cos)
plt.title("cosine curve")

# plt.show() # plt.show() mainly reqd in spider , in jupyter we can use inline to show plot,but will be changed for few cases
```

```python
## bar plot
x = [2, 8, 10]
y = [11, 16, 9] # y values will be frequency of x values

x2 = [3, 9, 11]
y2 = [6, 15, 7] # y2 will be frequency of x2

plt.bar(x,y)
plt.show()

plt.bar(x2, y2, color='g')
plt.title("Bar Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
## Histograms
a = np.array([22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
plt.hist(a) # default bin size = 10
plt.show()

plt.hist(a, bins=20)
plt.title("Histogram Example")
plt.show()
```

```python
## Box Plot
data = [np.random.normal(0, std, 100) for std in range(1,4)]

data # list of 3 arrays
```

```python
# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()

plt.boxplot(data, vert=False, patch_artist=False)
plt.show()
```

```python
## Pie Chart
labels = ["Python", "C++", "Ruby", "Java"] # can use without list 
sizes = [215, 130, 245, 210]
colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode = [0.4, 0, 0, 0] # explode 1st slice # can put in tuple
plt.subplot(1,2,1)
plt.pie(x = sizes, labels = labels, colors = colors, explode = explode, autopct='%1.1f%%', shadow=True)
# plt.show() 
# this plt.show() was killing subplot, if will keep this two different within different plots subplots will be created, 
# not in sa same plot two subplots will be created



labels = ["Python", "C++", "Ruby", "Java"] # can use without list 
sizes = [215, 130, 245, 210]
colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode = [0.4, 0, 0.1, 0] # explode 1st slice # can put in tuple
plt.subplot(1,2,2)
plt.pie(x = sizes, labels = labels, colors = colors, explode = explode, autopct='%1.3f%%', shadow=False)
plt.axis="equal"
plt.show()
```
