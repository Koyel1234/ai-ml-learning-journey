# NumPy

NumPy is general-purpose array-processing package. It provides high-performing multidimentional array object, and tools for working with these arrays. It is the fundmental package for scientific computing with python.

# What is an array?

An array is a data structure that stores values of same data type. In Python, this is the main difference between arrays and lists. While Python lists can obtain values with different data types, arrays in python can only contain values corresponding to same data type.

```python
# copy

a6 = np.array([1,2,3,500,500,500,500,500])
a7=a6 # sharing same memory location
print(a6, a7)
```

```python
a7[3:] = 1000
print(a6, a7) # both changed as memory loc is same
```

```python
a8 = np.array([1,2,3,500,500,500,500,500])
a9 = a8.copy() # copying a8 in different memory location
print(a8, a9)
a9[3:] = 1000
print(a8, a9) # only a9 will br changing as different memory locations are there
```
